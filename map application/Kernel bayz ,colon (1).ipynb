```python
import serial

# Define the serial port and baud rate
serial_port = 'COM5'  # Adjust this according to your device's serial port
baud_rate = 9600

# Initialize serial communication
ser = serial.Serial(serial_port, baud_rate)

route_points=[]
i=0
try:
    while True:
        # Read data from the serial port
        line = ser.readline().decode('utf-8').strip()
        print("the data which reterived: ",line)  
        points=line.split(',')
        route_points.append((float(points[0]),float(points[1])))
        print("the data which reterived: ",route_points) 



except KeyboardInterrupt:
    # Close the serial connection on Ctrl+C
    ser.close()
    print("Serial connection closed.")

# Print a message indicating successful data writing
print("Data has been written to the file successfully.")
```

### Importing Libraries 

```python
import folium
from folium.plugins import PolyLineTextPath
import matplotlib.pyplot as plt
```

### Starting Map 

```python

# Initialize a folium map at the first location
# Central latitude and longitude
central_latitude = 30.06515
central_longitude = 31.27905

# Generate points around the central location with small variations
route_points = [
    (central_latitude, central_longitude),
    (central_latitude + 0.0001, central_longitude + 0.0002),
    (central_latitude + 0.0002, central_longitude + 0.0001),
    (central_latitude + 0.0003, central_longitude - 0.0001),
    (central_latitude + 0.0004, central_longitude - 0.0002),
    (central_latitude + 0.0005, central_longitude - 0.0003),
    (central_latitude + 0.0004, central_longitude - 0.0004),
    (central_latitude + 0.0003, central_longitude - 0.0005),
    (central_latitude + 0.0002, central_longitude - 0.0004)
]
```

```python
initial_point = route_points[0]
map_display = folium.Map(location=initial_point, zoom_start=18)
```

### Ploting the points in the Map

```python


# Adding markers and lines along the route
folium.Marker(route_points[0], popup='Beginning', icon=folium.Icon(color='green')).add_to(map_display)
for idx in range(1,len(route_points)-1):
    folium.PolyLine(locations=[route_points[idx - 1], route_points[idx]], color='blue', dash_array='5, 5').add_to(map_display)
folium.Marker(route_points[len(route_points)-1], popup='Destination', icon=folium.Icon(color='red')).add_to(map_display)
        
```
